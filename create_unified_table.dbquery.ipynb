```sql
USE CATALOG workspace;
USE SCHEMA med;

-- combine tables avoiding structure misalignment
CREATE OR REPLACE TABLE raw_docs AS
SELECT
  doc_id,
  category,
  source,
  title,
  synonyms,
  url,
  raw_text,
  meta_json,
  ingested_at
FROM raw_docs_medlineplus

UNION ALL

SELECT
  doc_id,
  category,
  source,
  title,
  synonyms,
  url,
  raw_text,
  meta_json,
  ingested_at
FROM raw_docs_openfda;

-- verify union
SELECT
  (SELECT count(*) FROM raw_docs_medlineplus) as medlineplus_count,
  (SELECT count(*) FROM raw_docs_openfda) as openfda_count,
  (SELECT count(*) FROM raw_docs) as union_count;

DESCRIBE TABLE raw_docs;

SELECT * FROM raw_docs LIMIT 20;
```
