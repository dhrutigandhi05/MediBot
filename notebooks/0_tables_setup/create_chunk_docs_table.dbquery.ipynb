```sql
USE CATALOG workspace;
USE SCHEMA med;

CREATE TABLE IF NOT EXISTS doc_chunks (
  doc_id       STRING,
  source       STRING,
  chunk_id     STRING,
  chunk_index  INT,
  chunk_text   STRING,
  title        STRING,
  category     STRING,
  url          STRING,
  synonyms     ARRAY<STRING>,
  snapshot_ts  TIMESTAMP
)
USING DELTA;

SELECT source, category, count(*)
FROM doc_chunks
GROUP BY source, category;
```
