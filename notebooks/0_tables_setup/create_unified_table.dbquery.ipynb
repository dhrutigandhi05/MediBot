```sql
USE CATALOG workspace;
USE SCHEMA med;

CREATE TABLE IF NOT EXISTS raw_data (
  doc_id STRING,
  category STRING,
  source STRING,
  title STRING,
  synonyms ARRAY<STRING>,
  url STRING,
  raw_text STRING,
  meta_json STRING,
  ingested_at TIMESTAMP
)
USING DELTA;

DESCRIBE TABLE raw_data;

SELECT source, count(*)
FROM raw_data
GROUP BY source;

SELECT *
FROM raw_data
ORDER BY ingested_at DESC
LIMIT 20;
```
