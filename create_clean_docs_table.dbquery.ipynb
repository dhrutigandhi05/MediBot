```sql
USE CATALOG workspace;
USE SCHEMA med;

CREATE TABLE IF NOT EXISTS docs_clean (
  doc_id        STRING,
  source        STRING,
  category      STRING,
  title         STRING,
  synonyms      ARRAY<STRING>,
  url           STRING,
  text_clean    STRING,
  meta_json     STRING,
  snapshot_ts   TIMESTAMP
)
USING DELTA;

-- remove meta_json
REPLACE TABLE workspace.med.docs_clean AS
SELECT 
  doc_id,
  source,
  category,
  title,
  synonyms,
  url,
  text_clean,
  snapshot_ts
FROM workspace.med.docs_clean;

DESCRIBE workspace.med.docs_clean;
SELECT * FROM workspace.med.docs_clean;
```
